```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://applefans.today/category/news/'
pages_to_crawl = 10 # the pages you want to get; starts to count from zero
```

```python
def filter (title):
    #'title' is tag element type, titletext is string, need to convert before filtering
    titlestr = str(title)
    
    #-----FILTER SECTION START-----
    # in this case, we want to filter the articles that 
    # includes "iPhone" but not "iPhone 12"
    
    # returns "no" if the title is "iPhone 12"
    if 'iPhone 12' in titlestr:
        return 'no'
    # only return the title if it contains "iPhone"
    if 'iPhone' in titlestr:
        return title
    else:
        return 'no'
    #-----FILTER SECTION END-----
    # if you want to filter nothing (return every title from the page)
    # comment the FILTER SECTION code above and simply return title
#     return title
```

```python
for o in range (pages_to_crawl):

    r = requests.get(url)
    soup = BeautifulSoup(r.text,"html.parser")
    title = soup.find_all('h3', class_='title')
    # get the titles 
    for i in title:

        if '文章分類' in i.text:
            break
        if not filter(i.text) == 'no':
            
            print('Page:' + str(o)) # print out the page number
            print(i.text) # print out the title of the article
            print(i.a.get('href')) # print out the link to the article

    url = 'https://applefans.today/category/news/page/' + str(o+1) + '/' # switch to the next page
```
